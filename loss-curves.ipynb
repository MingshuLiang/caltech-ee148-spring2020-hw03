```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
set_por = [0.0625, 0.125, 0.25, 0.5, 1]
training_loss = [0.0038, 0.0030, 0.0024, 0.0017, 0.0013]
test_loss = [0.0553, 0.0361, 0.0237, 0.0182, 0.0135]
```

```python
plt.figure()
plt.loglog(set_por, training_loss, label = 'training loss')
plt.loglog(set_por, test_loss, label = 'test loss')
plt.xlabel('training set size')
plt.ylabel('loss')
plt.legend()
plt.title('Loss curves')
plt.show
plt.savefig('loss-curves.png')
```
